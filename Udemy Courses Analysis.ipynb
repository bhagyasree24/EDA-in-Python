```python
#creating price buckets to see the distribution of courses 
dict={'price50':0,'price100':0,'price200':0}
for price in df['price']:
    if price >0 and price <50:
        dict['price50']+=1
    if price >=50 and price <100:
        dict['price100']+=1
    if price >=100 and price <=200:
        dict['price200']+=1
```

```python
dict
```

```python
#plot of total courses in each price buckets
buckets=plt.bar(dict.keys(),dict.values(),color='#50763B')
for bucket in buckets.patches:
    bar_val=bucket.get_height()
    x_text=bucket.get_x() +  bucket.get_width()/2
    y_text=bucket.get_height() - bucket.get_height()/2
    plt.text(x_text,y_text,bar_val,fontsize=11,ha='center',color='w')
    plt.xticks([0,1,2],['Price from 1 to 50','Price from 50 to 100','Price from 100 to 200'])
    plt.title('Total Courses by Price Buckets')
```

```python
from wordcloud import WordCloud, STOPWORDS
```

### Find the most common keywords in Course Titles

```python

comment_words = ''
stopwords = set(STOPWORDS)
 
# iterate through course title 
for val in df.course_title:
     
    # typecaste each val to string
    val = str(val)
 
    # split the string
    tokens = val.split()
     
    # Converts each token into lowercase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
     
    comment_words += " ".join(tokens)+" "
WordCloud(
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)
 
plt.figure(figsize = (8,10) ,facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
```
